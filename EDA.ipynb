# **PRIMERO ANALIZAREMOS LOS MESES Y DIAS DONDE SE ESTRENARON LAS PELICULAS**

```python
# PARA ESO USAREMOS LA COLUMNA RELEASE_DATE
# Diccionario de nombres de meses 
meses_espanol = {
    1: 'enero',
    2: 'febrero',
    3: 'marzo',
    4: 'abril',
    5: 'mayo',
    6: 'junio',
    7: 'julio',
    8: 'agosto',
    9: 'septiembre',
    10: 'octubre',
    11: 'noviembre',
    12: 'diciembre'
}
# Convertir la columna "release_date" a formato de fecha
df['release_date'] = pd.to_datetime(df['release_date'])

# Extraer el mes y crear una nueva columna "meses"
df['meses'] = df['release_date'].dt.month

# Contar la cantidad de películas por mes
movies_por_mes = df['meses'].value_counts().sort_index()

# Ordenar los resultados por mes 
movies_por_mes = movies_por_mes.rename(index=meses_espanol)

movies_por_mes
```

### Hay podemos ver que el mes donde mas peliculas se estrenaron fue en enero

```python
# Ahora haremos uno para los dias
# diccionario de dias 
meses_espanol = {
    0: 'Lunes',
    1: 'Martes',
    2: 'Miercoles',
    3: 'Jueves',
    4: 'Viernes',
    5: 'Sabado',
    6: 'Domingo'
}
# Convertir la columna "release_date" a formato de fecha
df['release_date'] = pd.to_datetime(df['release_date'])

# Extraer el dia y crear una nueva columna "DIAS"
df['DIAS'] = df['release_date'].dt.dayofweek

# Contar la cantidad de películas por dias
movies_por_dia = df['DIAS'].value_counts().sort_index()

# Ordenar los resultados por dia
movies_por_dia = movies_por_dia.rename(index=meses_espanol)

movies_por_dia
```

### PODEMOS APRECIAR  QUE EL DIA QUE MAS PELICULAS SE VIERON FUE EL VIERNES

### PODEMOS CONCLUIR QUE NO HAY MULTICOLINEALIDAD ENTRE MIS VARIABLES QUE ME AYUDARAN A SELECCIONAR LAS PELICULAS PARA MI SISTEMA DE RECOMENDACIONES

# **AHORA ANALIZAREMOS LAS PELICULAS MAS VOTADAS Y CON MAYOR POPULARIDAD**

```python
# Ahora usaremos las columnas vote_count , vote_average y la de popularity

# primero vamos con la de vote_count
vote_counts=df['vote_count']
a = vote_counts.quantile(0.95)
a
```

```python
# segundo vamos con vote_average
vote_averages = df['vote_average']
b = vote_averages.mean()
b
```

```python
# con esto separamos las peliculas que seran parte de nuestra lista para el sistema de recomendaciones
calificadas = df[(df['vote_count'] >= a) & (df['vote_count'].notnull()) & (df['vote_average'].notnull())][['title', 'release_year', 'vote_count', 'vote_average', 'popularity', 'overview']]

calificadas.shape
```

```python
calificadas.head()
```

```python
calificadas.to_csv('calificadas.csv',index=False)
```
